### Initial imports

```python
%matplotlib inline
```

```python
import numpy as np
from numpy import linspace, vectorize, concatenate, pi, sin, cos
```

```python
import matplotlib.pyplot as plt
```

```python
import math
from math import sqrt
```

### Draw circle with two lines

```python
X = linspace(-1, 1, 360)

def f(x):
    return sqrt(1 - x**2)
f = vectorize(f)

plt.plot(X, f(X))
plt.plot(X, f(X) * -1)
plt.axis('equal')
```

### Draw circle with one line

```python
Xd = linspace(-1, 1, 360)
def f(x):
    return sqrt(1 - x**2)
f = vectorize(f)

X = concatenate([Xd, Xd*-1])
Y = concatenate([f(Xd), f(Xd*-1)*-1])

plt.plot(X, Y)
plt.axis('equal')
```

### Draw circle using trigonometry

```python
theta = linspace(0, 2*pi, 360)

plt.plot(cos(theta), sin(theta))
plt.axis('equal')
```

### Relationship between x & y coordinates of a circle and cos/sin

```python
a = 1.8 # Any number between 0 and 2
theta = linspace(0, a*pi, a*180)

plt.figure(figsize=(12,8))

plt.subplot(2, 2, 1)
plt.axis([-1.5, 1.5, -1, 1])
plt.axhline(color='black')
plt.axvline(color='black')
plt.plot(cos(theta), sin(theta)) # Draw arc
plt.plot([0, math.cos(a*pi)], [0, math.sin(a*pi)]) # Draw radius
plt.plot([math.cos(a*pi), 1.5], [math.sin(a*pi), math.sin(a*pi)], 'r') # Draw horizontal line
plt.plot([math.cos(a*pi), math.cos(a*pi)], [math.sin(a*pi), -1.0], 'r') # Draw vertical line

plt.subplot(2, 2, 2)
plt.axis([0, 7, -1, 1])
plt.axhline(color='black')
plt.plot(theta, sin(theta)) # Draw sin(x) curve
plt.plot([0, a*pi], [math.sin(a*pi), math.sin(a*pi)], 'r') # Draw horizontal line

plt.subplot(2, 2, 3)
plt.axis([-1.5, 1.5, 0, 7])
plt.gca().invert_yaxis()
plt.axvline(color='black')
plt.plot(cos(theta), theta) # Draw cos(x) curve
plt.plot([math.cos(a*pi), math.cos(a*pi)], [0, a*pi], 'r') # Draw vertical line
```
