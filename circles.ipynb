### Initial imports

```python
%matplotlib inline
```

```python
import numpy as np
from numpy import linspace, vectorize, concatenate, pi, sin, cos
```

```python
import matplotlib.pyplot as plt
```

```python
import math
from math import sqrt
```

### Draw circle with two lines

```python
X = linspace(-1, 1, 360)

def f(x):
    return sqrt(1 - x**2)
f = vectorize(f)

plt.plot(X, f(X))
plt.plot(X, f(X) * -1)
plt.axis('equal')
```

### Draw circle with one line

```python
Xd = linspace(-1, 1, 360)
def f(x):
    return sqrt(1 - x**2)
f = vectorize(f)

X = concatenate([Xd, Xd*-1])
Y = concatenate([f(Xd), f(Xd*-1)*-1])

plt.plot(X, Y)
plt.axis('equal')
```

### Draw circle using trigonometry

```python
theta = linspace(0, 2*pi, 360)

plt.plot(cos(theta), sin(theta))
plt.axis('equal')
```

### Relationship between x & y coordinates of a circle and cos/sin

```python
a = 1.8 # Any number between 0 and 2
end_theta = a*pi
end_x, end_y = math.cos(end_theta), math.sin(end_theta)
theta = linspace(0, a*pi, a*180)

plt.figure(figsize=(12,8))
plt.axhline(color='black')
plt.axvline(color='black')

plt.subplot(2, 2, 1)
plt.title('unit circle')
plt.axis([-1.5, 1.5, -1, 1])
plt.axhline(color='gray')
plt.axvline(color='gray')
plt.plot(cos(theta), sin(theta)) # Draw arc
plt.plot([0, end_x], [0, end_y]) # Draw radius
plt.plot([end_x, 1.5], [end_y, end_y], 'r') # Draw horizontal line
plt.plot([end_x, end_x], [end_y, -1.0], 'r') # Draw vertical line

plt.subplot(2, 2, 2)
plt.title(r'sin($\theta$)')
plt.axis([0, 7, -1, 1])
plt.axhline(color='gray')
plt.plot(theta, sin(theta)) # Draw sin(x) curve
plt.plot([0, end_theta], [end_y, end_y], 'r') # Draw horizontal line

plt.subplot(2, 2, 3)
plt.title(r'cos($\theta$)')
plt.axis([-1.5, 1.5, 0, 7])
plt.axvline(color='gray')
plt.gca().invert_yaxis()
plt.plot(cos(theta), theta) # Draw cos(x) curve
plt.plot([end_x, end_x], [0, end_theta], 'r') # Draw vertical line
```
