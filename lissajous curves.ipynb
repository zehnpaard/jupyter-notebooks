# Exploration of Lissajous Curves
## Inspired by "Math Girls Talk about Trigonometry" by Hiro Yuki

### Initial imports

```python
%matplotlib inline
```

```python
import numpy as np
from numpy import linspace, pi, sin, cos
```

```python
import matplotlib.pyplot as plt
```

### Define $\theta$ as 360 equally distanced values between 0 and 2$\pi$

```python
theta = linspace(0, 2*pi, 360)
```

### Plotting (cos($\theta$), sin($\theta$ + b)) for b = 0, $\pi$/6, $\pi$/3, $\pi$/2 ... 11$\pi$/6, 2$\pi$

```python
for i in range(13):
    plt.subplot(4, 4, i+1)
    plt.plot(cos(theta), sin(theta + (i * pi/6)))
```

### Plotting (cos($\theta$), sin(2$\theta$ + b)) for b = 0, $\pi$/6, $\pi$/3, $\pi$/2 ... 11$\pi$/6, 2$\pi$

```python
for i in range(13):
    plt.subplot(4, 4, 1+i)
    plt.plot(cos(theta), sin(2*theta + (i * pi/6)))
    plt.axis('equal')
    plt.grid()
```

### Plotting (cos($\theta$), sin(3$\theta$ + b)) for b = 0, $\pi$/6, $\pi$/3, $\pi$/2 ... 11$\pi$/6, 2$\pi$

```python
for i in range(13):
    plt.subplot(4, 4, 1+i)
    plt.plot(cos(theta), sin(3*theta + (i * pi/6)))
    plt.axis('equal')
    plt.grid()
```

### Plotting (cos(2$\theta$), sin(3$\theta$ + b)) for b = 0, $\pi$/6, $\pi$/3, $\pi$/2 ... 11$\pi$/6, 2$\pi$

```python
for i in range(13):
    plt.subplot(4, 4, 1+i)
    plt.plot(cos(2*theta), sin(3*theta + (i * pi/6)))
    plt.axis('equal')
    plt.grid()
```

### Plotting (cos(3$\theta$), sin(2$\theta$ + b)) for b = 0, $\pi$/6, $\pi$/3, $\pi$/2 ... 11$\pi$/6, 2$\pi$

```python
for i in range(13):
    plt.subplot(4, 4, 1+i)
    plt.plot(cos(3*theta), sin(2*theta + (i * pi/6)))
    plt.axis('equal')
    plt.grid()
```
